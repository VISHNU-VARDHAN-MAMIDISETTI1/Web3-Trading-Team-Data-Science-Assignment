```python
import pandas as pd
df_fg = pd.read_csv("../csv/fear_greed_index.csv")

df_fg.columns = [c.strip() for c in df_fg.columns]
date_col = df_fg.columns[0]
```

```python
df_fg.isnull().sum()
```

```python

df_fg[date_col] = pd.to_datetime(df_fg[date_col], errors='coerce')
df_fg = df_fg.rename(columns={date_col: "Date"})
```

```python
df_fg
```

```python
df_fg.shape
```

```python
if 'Score' not in df_fg.columns:
    score_col = [c for c in df_fg.columns if c.lower() not in ('date', 'classification')]
    if score_col:
        df_fg['Score'] = pd.to_numeric(df_fg[score_col[0]], errors='coerce')
```

```python
if 'Classification' not in df_fg.columns and 'Score' in df_fg.columns:
    def label(s):
        if s < 25: return "Extreme Fear"
        elif s < 40: return "Fear"
        elif s < 60: return "Neutral"
        elif s < 75: return "Greed"
        else: return "Extreme Greed"
    df_fg['Classification'] = df_fg['Score'].apply(label)
```

```python
df_fg = df_fg[['Date', 'Score', 'Classification']].dropna(subset=['Date'])
```

```python
df_tr['Timestamp IST'] = pd.to_datetime(df_tr['Timestamp IST'], errors='coerce', dayfirst=True)
df_tr['Date'] = df_tr['Timestamp IST'].dt.floor('d')
```

```python
df_tr['Size USD'] = pd.to_numeric(df_tr['Size USD'], errors='coerce')
df_tr['Closed PnL'] = pd.to_numeric(df_tr['Closed PnL'], errors='coerce')
```

```python
daily_tr = df_tr.groupby('Date').agg(
    trades_count=('Account', 'count'),
    total_volume_usd=('Size USD', 'sum'),
    avg_closed_pnl=('Closed PnL', 'mean'),
    positive_trades=('Closed PnL', lambda x: (x > 0).sum())
).reset_index()
daily_tr['win_rate'] = daily_tr['positive_trades'] / daily_tr['trades_count']

```

```python
df_merge = pd.merge(df_fg, daily_tr, on='Date', how='left')

# Save combined CSV
merged_path = "../csv/merged_fg_trader.csv"
df_merge.to_csv(merged_path, index=False)
merged_path
```
